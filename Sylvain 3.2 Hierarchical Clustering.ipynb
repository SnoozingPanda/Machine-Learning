# Hierarchical Clustering
Within this, we're going to use Scikit-Learn and Matplotlib to visualize how hierarchical clustering works.

```
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
from sklearn.cluster import AgglomerativeClustering

data = pd.read_csv('https://gist.githubusercontent.com/keeganhines/59974f1ebef97bbaa44fb19143f90bad/raw/d9bcf657f97201394a59fffd801c44347eb7e28d/Hitters.csv')
```

```
data.head()
```

```
# Remove the several columns for the sake of this exercise
data = data.drop(['Unnamed: 0', 'League', 'Division', 'NewLeague'], axis=1)

# Remove rows with NaN
# INSERT CODE HERE
df_sans_nan = data.dropna()

# View the first five rows
# INSERT CODE HERE
print(df_sans_nan.head().to_markdown())
```

```
# Only use the first 50 datapoints of the dataset for easy visualization
# INSERT CODE HERE
df_head = df_sans_nan.head(50)
```

```
# View the shape of the data frame. It should be (50, 17)
# INSERT CODE HERE
df_head.shape
```

```
# n_clusters = int or None, default=2
# linkage = {â€˜wardâ€™, â€˜completeâ€™, â€˜averageâ€™, â€˜singleâ€™}, default=â€™wardâ€™
# distance_threshold = float, default=None
```

```
# Using the AgglomerativeClustering() function with a distance threshold of 0 and no n_clusters, fit the dataset.
# INSERT CODE HERE

clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=0).fit(df_head)

clustering_model
clustering_model.labels_
```

```
# Print out the distances of that model
# I.e. n_clusters_, min(), and max()
# INSERT CODE HERE
print(clustering_model.n_clusters_)

# there is no min() or max() for clustering models
```

```
#linkage possibilities = {â€˜wardâ€™, â€˜completeâ€™, â€˜averageâ€™, â€˜singleâ€™} => no centroid
```

#Single linkage

```
# Single
# Using single linkage, create and display a dendogram.
# INSERT CODE HERE
clustering_model_single = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage = "single" ).fit(df_head)

clustering_model_single
```

```
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 7))
plt.title("DF HEAD DENDOGRAM SINGLE")

selected_data = df_head
clusters = shc.linkage(selected_data,
            method='single',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

#Complete linkage

```
# Complete
# Using complete linkage, create and display a dendogram.
# INSERT CODE HERE
clustering_model_complete = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage = "complete" ).fit(df_head)

clustering_model_complete
```

```
plt.figure(figsize=(10, 7))
plt.title("DF HEAD DENDOGRAM COMPLETE")

selected_data = df_head
clusters = shc.linkage(selected_data,
            method='complete',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

#Average linkage

```
# Average
# Using average linkage, create and display a dendogram.
# INSERT CODE HERE
clustering_model_average = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage = "average" ).fit(df_head)

clustering_model_average
```

```
plt.figure(figsize=(10, 7))
plt.title("DF HEAD DENDOGRAM AVERAGE")

selected_data = df_head
clusters = shc.linkage(selected_data,
            method='average',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

#Ward linkage

```
# Centroid
# Using centroid linkage, create and display a dendogram.
# INSERT CODE HERE

# there is no centroid linkage in that model so I will use ward linkage instead
clustering_model_ward = AgglomerativeClustering(n_clusters=None, distance_threshold=0, linkage = "ward" ).fit(df_head)

clustering_model_ward
```

```
plt.figure(figsize=(10, 7))
plt.title("DF HEAD DENDOGRAM WARD")

selected_data = df_head
clusters = shc.linkage(selected_data,
            method='ward',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```

```
# After observing the dendograms above, let's say that you want to choose "Average" linkage with only 2 clusters.
# What are the labels of each of the datapoints?
# Print these labels out
# INSERT CODE HERE
clustering_model_average_2 = AgglomerativeClustering(n_clusters=2, distance_threshold=None, linkage = "average" ).fit(df_head)

clustering_model_average_2
```

```
plt.figure(figsize=(10, 7))
plt.title("DF HEAD DENDOGRAM AVERAGE 2 CLUSTERS")

selected_data = df_head
clusters = shc.linkage(selected_data,
            method='average',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.show()
```
