#Decision Tree Classifier

```python
X = df[["PassengerId", "Pclass", "Age", "SibSp", "Parch", "Fare"]]
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, train_size = 0.75)
print("The length of the initial dataset is :", len(X))
print("The length of the train dataset is   :", len(X_train))
print("The length of the test dataset is    :", len(X_test))
```

```python
from sklearn.tree import DecisionTreeClassifier

modelDTC = DecisionTreeClassifier()
modelDTC.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

x_pred = modelDTC.predict(X_train)
y_pred = modelDTC.predict(X_test)

accuracy1 = accuracy_score(y_train, x_pred)
accuracy2 = accuracy_score(y_test, y_pred)
print(f"Accuracy on train: {round(accuracy1, 2)}")
print(f"Accuracy on test: {round(accuracy2, 2)}")
```

#GridSearch

```python
from sklearn.model_selection import GridSearchCV
dico = {"max_depth": range(1, 51), "min_samples_leaf" : range(1, 16), "min_samples_split" : [2, 5, 7, 10, 15, 30]}
grid = GridSearchCV(DecisionTreeClassifier(), dico )
grid.fit(X,y)
```

```python
print("best score:",grid.best_score_)
print("best parameters:",grid.best_params_)
```

```python
grid.cv_results_
```

#RandomSearch

```python
from sklearn.model_selection import RandomizedSearchCV
dico = {"max_depth": range(1, 51), "min_samples_leaf" : range(1, 16), "min_samples_split" : [2, 5, 7, 10, 15, 30]}
rando = RandomizedSearchCV(DecisionTreeClassifier(), dico ,
                           n_iter=200)
rando.fit(X,y)
```

```python
print("best score:", rando.best_score_)
print("best parameters:", rando.best_params_)
```

```python
rando.cv_results_
```

Ici on constate que les scores de grid et random sont très différents pour min_samples_split (2 pour grid et 7 pour random) et min_samples_leaf (3 pour grid et 6 pour random)

En revanche ils sont proches pour max_depth (2 pour grid, 3 pour random)

On constate également que les scores des deux méthodes sont relativement proches

0.71 pour grid et 0.70 pour random

Dans le cas présent, le fait de tester les paramètres sur tout le dataset ou sur des éléments aléatoires ne fait pas grande différence
